## **Condicionales Anidados**

Es un condicional dentro de otro condicional

```python
numero =float(input("señor usuario capture el numero: "))
if numero > 0:
    print(f" el numero {numero} es positivo")
else:
    if numero == 0 :
        print(f"el numero {numero} es cero y es neutro")
    else:
      print(f"el numero {numero} es negativo")
```

```python
i = 1 #inicialización de la variable de control
while i <= 10: #la condición del ciclo
    print(i)
    i = i + 1 #cambio en la variable de control
```
